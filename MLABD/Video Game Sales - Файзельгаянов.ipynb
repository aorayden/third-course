#### Описание данных

Этот набор данных содержит список видеоигр, продажи которых превысили 100 000 копий.

Name - Название игры

Platform - Платформа выпуска игр (например, PC, PS4 и т.д.)

Year - Год выпуска игры

Genre - Жанр игры

Publisher - Издатель игры

NA_Sales - Продажи в Северной Америке (в миллионах)

EU_Sales - Продажи в Европе (в миллионах)

JP_Sales - Продажи в Японии (в миллионах)

Other_Sales - Продажи в остальном мире (в миллионах)

Global_Sales - Общий объем продаж по всему миру.

#### 9. Лучшие игры за 5 лет выпуска по жанрам

```python
years = [2007, 2008, 2009, 2010, 2011]
df_top = vgs[vgs['Year'].isin(years)]
df_count = df_top.groupby(['Year', 'Genre']).size().reset_index(name='count')
plt.figure(figsize=(18, 8))
sns.barplot(data=df_count, x='Year', y='count', hue='Genre')
plt.xlabel('Year')
plt.ylabel('count')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### 11. Игры какого жанра имеют самую высокую цену продажи в мире

```python
genre_sales = vgs.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=genre_sales, x='Genre', y='Global_Sales')
plt.xlabel('Genre')
plt.ylabel('Global_Sales')
plt.xticks(rotation=90)
plt.show()
```

#### 12. На какой платформе самая высокая цена продажи в мире?

```python
platform_sales = vgs.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(16, 8))
sns.barplot(data=platform_sales, x='Platform', y='Global_Sales')
plt.xlabel('Platform')
plt.ylabel('Global_Sales')
plt.xticks(rotation=90)
plt.show()
```

#### 13. Выведите топ 20 Publisher

#### 14. Общий доход по регионам

```python
vgs
```

```python
sales_sum = vgs[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
df_sales = sales_sum.reset_index()
df_sales.columns = ['region', 'sale']
print(df_sales)
```

```python
plt.figure(figsize=(7, 7))
plt.pie(df_sales['sale'], labels=df_sales['region'], colors=['#1976d2', '#ff9800', '#43a047', '#d32f2f'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df_sales, x='region', y='sale')
plt.xlabel('region')
plt.ylabel('sale')
plt.show()
```

```python
corr = vgs[['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr()
plt.figure(figsize=(9, 9))
sns.heatmap(corr, annot=True, cmap='Blues', vmin=-1, vmax=1)
plt.show()
```
